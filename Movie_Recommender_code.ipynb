```python
import pandas as pd
import numpy as np
import sklearn
df = pd.read_csv("mymoviedb.csv", engine="python")
```

```python
df = df[["Title", "Overview", "Genre", "Poster_Url"]]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df["Genre"] = df["Genre"].str.replace(" ","")
```

```python
df["Overview"] = df["Overview"].str.split()
```

```python
df["Genre"] = df["Genre"].str.replace(","," ").str.split()
```

```python
df["tag"] = df["Overview"] + df["Genre"]
```

```python
df["tag"].values
```

```python
new_df = df[["Title", "tag"]]
```

```python
new_df
```

```python
new_df["tag"] = new_df["tag"].apply(lambda x: " ".join(x))  # Converting list attributes into strings again
```

```python
new_df["tag"][0]
```

```python
new_df["tag"] = new_df["tag"].str.lower() #Lower Case conversion on=f strings
```

```python
new_df.head()
```

# Stemming : <h4>Deducing the words to its generalized form which prevents repition of words</h4>

```python
# Performing stemming on TAG column

from nltk.stem import PorterStemmer

porter = PorterStemmer()

l = []

for i in new_df["tag"]:
    l.append(porter.stem(i))
new_df["tag"] = l
```

```python
new_df.head()
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words=None)
```

```python
vectors = cv.fit_transform(new_df["tag"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    mov_ind = new_df[new_df["Title"]==movie].index[0]
    distances = similarity[mov_ind]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]
    for i in movies_list:
        print(new_df.iloc[i[0]].Title)
```

```python
recommend("Spider-Man: No Way Home")
```
